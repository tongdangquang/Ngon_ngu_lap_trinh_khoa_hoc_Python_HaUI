```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Yêu cầu 1.
def Nhap_n(): # hàm nhập n
    while True:
        n = input("Nhập một số nguyên dương n: ")
        if (not n.isdigit()) or (int(n) < 0):
            print("Bạn phải nhập đúng số nguyên dương!")
            continue
        return int(n)

def Nhap_mang(n): # hàm nhập mảng với n phần tử
    while True:
        a = list(map(float, input(f"Nhập số {n} cách nhau bởi dấu ',': ").split(",")))
        if len(a) == n:
            break
        else:
            print(f"Vui lòng nhập lại đúng {n} số thực!")
            continue
    a = np.array(a)
    return a

n = Nhap_n()
a = Nhap_mang(n)
b = Nhap_mang(n)
print(n)
print(a)
print(b)
```

```python
#  Yêu cầu 2.
def yeucau_2(a):
    try:
        k = Nhap_n()
        a_reshape = a.reshape((k, -1))
        return a_reshape
    except:
        return []

c = yeucau_2(a)
if (len(c) > 0):
    print(f"Mảng a chuyển thành 2 chiều:\n{c}")
else:
    print("Không thể chuyển a thành mảng 2 chiều vì kích thước không hợp lệ!")
```

```python
# Yêu cầu 3.
def yeucau_3(a, f, t):
    x = a[np.where((f <= a) & (a <= t))]
    x_sum = np.sum(x)
    return x, x_sum

x, x_sum = yeucau_3(b, 2, 5)
if len(x) > 0:
    print(f"Các giá trị thuộc khoảng [2, 5] trong mảng b là:\n{x}")
    print(f"Tổng của chúng là: {x_sum}")
else:
    print("Không có giá trị nào thỏa mãn điều kiện!")
```

```python
# Yêu cầu 4:
def yeucau_4(a, b):
    a = np.sort(a)
    b = np.sort(b)[::-1]
    c = np.concatenate((a, b))
    return c

print(f"Mảng được ghép từ a và b là:\n{yeucau_4(a, b)}")
```

```python
# Yêu cầu 5.
def pie_chart(a):
    plt.figure(figsize=(10, 10))
    wp = {"linewidth": 1, "edgecolor": "yellow"}
    plt.pie(a, explode=[0.05 for i in range(len(a))], labels=a, wedgeprops=wp, shadow=True, autopct='%.2f')
    plt.title("Pie chart for a array", fontweight = "bold", fontsize = 20)
    plt.show()
pie_chart(a)
    
def plot_chart(a, b):
    plt.figure(figsize=(10, 10))
    plt.plot(a, b, marker = "*", color = "blue")
    plt.title("Plot chart for a array", fontweight = "bold", fontsize = 20)
    plt.xlabel("Trục x (a)", fontweight = "bold")
    plt.ylabel("Trục y (b)", fontweight = "bold")
    plt.grid(linestyle = ":", linewidth = 1)
    plt.show()

plot_chart(a, b)
```
