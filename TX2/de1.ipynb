```python
import numpy as np
import pandas as ppd
import matplotlib.pyplot as plt
```

```python
# Yêu cầu 1.
def Nhap(): # hàm nhập dữ liệu
    while True:
        try:
            value = list(map(float, input("Nhập lần lượt a, b, c, d cách nhau 1 dấu phẩy: ").split(",")))
            if (len(value) != 4):
                print("Bạn phải nhập đủ 4 giá trị, vui lòng nhập lại!")
                continue
            a, b, c, d = value
            return a, b, c, d
        except:
            print("Bạn phải nhập đúng dữ liệu, vui lòng nhập lại!")

a, b, c, d = Nhap()
print(a, b, c, d)
```

```python
# Yêu cầu 2.
def Draw_chart(a, b, c, d): # vẽ đồ thị hàm số
    x = np.linspace(-20, 20, 2000)
    y = a*x**3 + b*x**2 + c*x + d

    plt.figure(figsize=(10, 8))
    plt.plot(x, y, marker="*", color = "blue")
    plt.title(f"Đồ thị hàm số {a}x3 + {b}x2 + {c}x + {d}")
    plt.xlabel("Trục x", loc="center")
    plt.ylabel("Trục y", loc="center")
    plt.axhline(color = "black", linewidth = 1)
    plt.axvline(color = "black", linewidth = 1)
    plt.grid(linestyle = ":", linewidth = 0.5)
    plt.show()

Draw_chart(a, b, c, d)
```

```python
# Yêu cầu 3.
def Nhap_mang(): # hàm nhập mảng
    while True:
        try:
            n = int(input("Nhập n = "))
            a = list(map(int, input(f"Nhập {n} phần tử cách nhau bởi dấu phẩy: ").split(",")))
            if (len(a) != n):
                print(f"Bạn phải nhấp đúng {n} giá trị!")
                continue
            a = np.array(a)
            return a, n    
        except:
            print("Vui lòng nhập lại cho đúng kiểu dữ liệu được yêu cầu!")

a, n = Nhap_mang()
print(f"Số lượng phần tử trong a: n = {n}")
print(f"Mảng a: a = {a}")
```

```python
# Yêu cầu 5. 
def yeucau_5(a):
    a_sorted = np.sort(a, kind="heapsort")[::-1]
    print(f"Mảng a sau khi sắp xếp giảm dần bằng heapsort: {a_sorted}")

yeucau_5(a)
```

```python
# Yêu cầu 6.
def yeucau_6(a):
    try:
        while True:
            p = input("Nhập một số nguyên dương: ").strip()
            if not p.isdigit():
                print("Bạn phải nhập một số nguyên dương!")
                continue
            p = int(p)
            if (p <= 0):
                print("Bạn phải nhập một số nguyên dương!")
                continue
            break
        a_reshape = a.reshape((p, -1))
        return a_reshape
    except:
        return []

a_reshape = yeucau_6(a)
if (len(a_reshape) == 0):
    print("Không thể chuyển mảng a thành mảng 2 chiều!")
else:
    print(f"Mảng 2 chiều:\n{a_reshape}")
```
